```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xlrd
import xlwt
import operator
import numpy as np
import array 
import operator
import datetime
from datetime import timedelta
test= pd.read_excel(r'C:\Users\49173\Desktop\energy.xlsx')
filled =test.fillna(-1)
StartOfJoruney=0
EndOfJoruney=0
CountOfJourneies=0
Start =False
End =True
filled =test.fillna(-1)
ArrayOfMeans= []
ArrayOfSums= []
ArrayOfDuration=[]
StartOfJoruney=0
EndOfJoruney=0
CountOfJourneies=0
slcied = filled.iloc[:,1:55]  
data=pd.DataFrame()
```

```python
for col in slcied.columns:
    for x in range(4,len(test[col])-1 ):
        ValueOfCell=slcied.loc[[x],[col]]   
        ValueOfNextCell=slcied.loc[[x+1],[col]] 
        if np.any(ValueOfCell>= 0) and bool(End)==True:
            Start=True
            End=False
            CountOfJourneies=CountOfJourneies+1
            StartOfJoruney=x
            StartDate=test.loc[x,'Datum']
        if np.any(ValueOfNextCell< 0)==True and bool(Start)==True :
            EndOfJoruney=x
            EndDate=test.loc[x,'Datum']
            Duration=EndDate-StartDate
            minutes=Duration.total_seconds()/60
            ArrayOfDuration.append(minutes)
            MeanEnergyOfJourney=test.loc[StartOfJoruney:EndOfJoruney,col].mean()
            SumEnergyOfJourney=test.loc[StartOfJoruney:EndOfJoruney,col].sum()
            ArrayOfMeans.append(MeanEnergyOfJourney)
            ArrayOfSums.append(SumEnergyOfJourney)
            StartOfJoruney=0
            EndOfJoruney=0
            Start=False 
            End=True 
    print("End of journies of the train")
    print("journes" )
    print(CountOfJourneies)
    CountOfJourneies=0
    plt.plot(ArrayOfMeans,ArrayOfDuration, 'ro')
    plt.ylabel('Duration in minutes')
    plt.xlabel('Mean enegy in journey by Kwh')
    plt.title(col)
    plt.show()
    plt.plot(ArrayOfSums,ArrayOfDuration, 'ro')
    plt.ylabel('Duration in minutes')
    plt.xlabel('total enegy in journey by Kwh')
    plt.title(col)
    plt.show()
    ArrayOfMeans.clear()
    ArrayOfSums.clear()
    ArrayOfDuration.clear()
```
